```python
import pandas as pd
import matplotlib.pyplot as plt

file = 'TEST CASE 4.csv'
df = pd.read_csv(file)
```

```python
plt.figure(figsize=(20, 10), dpi=150)
plt.plot(df.index, df.Food/100, label = ' Food')
plt.plot(df.index, df.Shelter/100, label = ' Shelter')
plt.plot(df.index, df.Household_operations_furnishings_and_equipment/100, label = ' Household Operations')
plt.plot(df.index, df.Clothing_and_footwear/100, label = ' Clothing')
plt.plot(df.index, df.Transportation/100, label = ' Transportation')
plt.plot(df.index, df.Health_and_personal_care/100, label = ' Healthcare')
plt.plot(df.index, df.Recreation_education_and_reading/100, label = ' Education')
plt.plot(df.index, df.Alcoholic_beverages_tobacco_products_and_recreational_cannabis/100, label = ' Alcohol/Tobacco/Cannabis')
plt.plot(df.index, df.All_items_excluding_food/100, label = ' Non-Food items')
plt.plot(df.index, df.All_items_excluding_food_and_energy/100, label = ' Non-Food and Non-Energy items')

plt.vlines(x=26, ymin=0.81, ymax=1.8, colors='black', ls='-', lw=1.5, label='WHO Issues Global Health Emergency')
plt.vlines(x=31.56, ymin=0.81, ymax=1.8, colors='black', ls='-.', lw=1.5, label='End of 1st wave')
plt.vlines(x=42.87, ymin=0.81, ymax=1.8, colors='black', ls='--', lw=1.5, label='End of 2nd wave ')

plt.gca().set(title='Percentage change in CPI between 2018 and 2021', 
xlabel='Months', ylabel='CPI')
plt.legend(bbox_to_anchor=(1.0, 1), loc='upper left')
plt.show()
```

***Cumulative total***

```python
plt.figure(figsize=(20, 10), dpi=150)
plt.plot(df.index, df.Cumulative_Total/100, label = 'Cumulative Total')

plt.vlines(x=26, ymin=1.3, ymax=1.45, colors='black', ls='-', lw=1.5, label='WHO Issues Global Health Emergency')
plt.vlines(x=31.56, ymin=1.3, ymax=1.45, colors='black', ls='-.', lw=1.5, label='End of 1st wave')
plt.vlines(x=42.87, ymin=1.3, ymax=1.45, colors='black', ls='--', lw=1.5, label='End of 2nd wave ')

plt.gca().set(title='Percentage growth in CPI between 2018 and 2021', 
xlabel='Months', ylabel='CPI')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file2 = 'TEST CASE 5 - Sheet1.csv'
df2 = pd.read_csv(file2)
```

```python
plt.figure(figsize=(20, 10), dpi=150)
plt.plot(df2.index, df2.Food, label = 'Food')

plt.gca().set(title='Monthly adjusted consumer expenditure basket weights (Food) during the pandemic (Feb 2020 to Jan 2021)', 
xlabel='Months', ylabel='Basket weight %')
plt.legend()
plt.show()
```
